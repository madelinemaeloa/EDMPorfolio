## Import, Clean, Prepare

```python
import pandas as pd
```

```python
# imports the crime dataset
crimeData = pd.read_csv("Crime_2015.csv")
crimeData
```

```python
# creates a new column called TotalCrime that sums all the various crime in each city
crimeData['TotalCrime'] = crimeData['ViolentCrime'].str.replace(',', "").astype(float) + crimeData['Murder'].astype(float) + crimeData['Rape'].astype(float) + crimeData['Robbery'].astype(float) + crimeData['AggravatedAssault'].astype(float) + crimeData['PropertyCrime'].str.replace(',', "").astype(float) + crimeData['Burglary'].str.replace(',', "").astype(float) + crimeData['Theft'].str.replace(',', "").astype(float) + crimeData['MotorVehicleTheft'].astype(float) 
crimeData
```

```python
# returns a list of all cities and its corresponding TotalCrime rate 
cityCrimeRate = crimeData[['CITY', 'TotalCrime']]
cityCrimeRate
```

```python
# returns only the cities with a TotalCrime score of less than 5000
lowCrimeCities = (cityCrimeRate['TotalCrime'] < 5000)
lowCrimeCities = cityCrimeRate[lowCrimeCities]
lowCrimeCities
```

```python
# imports collegeScorecard dataset
collegeScoreCard = pd.read_csv("collegeScorecard.csv",low_memory=False)
collegeScoreCard
```

```python
# creates a new table called data with the required columns only
data = collegeScoreCard[['INSTNM','CITY','STABBR','LOCALE','UGDS_WHITE','UGDS_BLACK','UGDS_HISP','UGDS_ASIAN','UGDS_AIAN','UGDS_NHPI','ADM_RATE_ALL']]
data
```

```python
# creates a new table called goodSchools with the required criteria for each column variable
goodSchools = (data['LOCALE'] > 11) & (data['UGDS_WHITE'] != 0) & (data['UGDS_BLACK'] != 0) & (data['UGDS_HISP'] != 0) & (data['UGDS_ASIAN'] != 0) & (data['UGDS_AIAN'] != 0) & (data['UGDS_NHPI'] != 0) & (data['ADM_RATE_ALL'] < 0.5)
goodSchools = data[goodSchools]
goodSchools
```

```python
# cleaned up table called goodSchoolsList with only the school name, city, and state
goodSchoolsList = goodSchools[['INSTNM','CITY','STABBR']]
goodSchoolsList
```

## Merge

```python
# merging the lowCrimeCities table and goodSchoolsList 
# shows the schools with crime rate < 5000 and with the required criteria 
finalList = pd.merge(lowCrimeCities, goodSchoolsList)
finalList
```

## Final List

```python
# cleaned up final list to display institution first, followed by city, state, then crime rate
finalList = finalList[["INSTNM", "CITY", "STABBR","TotalCrime"]]
finalList
```
